```python
q_value = np.random.rand(100,10)

discount_factor = 0.7

ratio = 0.9

for episode in range(EPISODE_NUM):
    
  location = 0

  old_state = 0
    
  next_state = 0

  fuel = 299
    
    
  while not is_terminate(fuel, next_state):
    
    index = np.argmax(q_value[next_state])
    
    temp = action_selection[index]
    
    action_list.append(temp)
    
    velocity, theta = action[temp]
    
    resistance = state_resistance(v)

    old_location = location
    
    old_state = sr(old_location)
    
    location = move(old_location, v, theta, resistance)
    
    next_state = sr(location)
    
    fuel = fuel - 10

    current_reward = get_reward(fuel, next_state)
    
    oqvalue = q_value[old_state, index]
    
    Temporal_difference = current_reward + (DISCOUNT * np.max(q_value[next_state])) - oqvalue

    new_q = oqvalue + ratio * Temporal_difference
    
    q_value[old_state, index] = new_q
    
  print(location)
```
